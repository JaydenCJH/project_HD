```python
Q1 = data[['PhysicalHealthDays', 'MentalHealthDays', 'SleepHours', 'HeightInMeters', 'WeightInKilograms', 'BMI']].quantile(q = 0.25)
Q3 = data[['PhysicalHealthDays', 'MentalHealthDays', 'SleepHours', 'HeightInMeters', 'WeightInKilograms', 'BMI']].quantile(q = 0.75)
print(Q1)
print()
print(Q3)
```

```python
IQR = Q3-Q1
print(IQR)
```

```python
print(Q1-IQR*1.5)
print()
print(Q3+1.5*IQR)
```

```python
data_IQR = data[(data['SleepHours'] <= Q3['SleepHours']+IQR['SleepHours']*1.5)&(data['SleepHours'] >= Q1['SleepHours']-IQR['SleepHours']*1.5)]
data_IQR = data_IQR[(data_IQR['BMI'] <= Q3['BMI']+IQR['BMI']*1.5)&(data_IQR['BMI'] >= Q1['BMI']-IQR['BMI']*1.5)]
```

```python
from scipy import stats
```

```python
z = np.abs(stats.zscore(data.loc[:, ['SleepHours', 'BMI']]))
threshold = 3
data_z = data[(z<threshold).all(axis = 1)]
```

```python
print('Original:', len(data))
print('IQR:', len(data_IQR))
print('zscore:', len(data_z))
```

```python
from sklearn.preprocessing import LabelEncoder

data = pd.get_dummies(data, columns = ['RaceEthnicityCategory', 'TetanusLast10Tdap'])
le = LabelEncoder()
for columns in data.select_dtypes(include = ['object', 'bool']).columns:
    data[columns] = le.fit_transform(data[columns])
```

```python
from sklearn import datasets
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
```

```python
x = data.drop(columns=['HadHeartAttack'])
y = data['HadHeartAttack']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3,random_state=42)
```

```python
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
classifier = LogisticRegression(random_state = 0)
classifier.fit(x_train, y_train)

y_pred = classifier.predict(x_test)
print ("Accuracy : ", accuracy_score(y_test, y_pred))
```

```python
sv = SVC(kernel='linear')
sv.fit(x_train, y_train)
```

```python
train_score= sv.score(x_train, y_train)
print("Training Score:", train_score)
```

```python
test_score= sv.score(x_test, y_test)
print("Testing Score:", test_score)
```

```python
knn = KNeighborsClassifier(n_neighbors=8)
knn.fit(x_train, y_train)

y_pred = knn.predict(x_test)
print(accuracy_score(y_test, y_pred))
```
