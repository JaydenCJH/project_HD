```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

pd.set_option('display.max_columns', 50)
```

```python
dataset = pd.read_csv("Heart_AfterPrep_0228.csv")
print(dataset.shape)
dataset.head()
```

```python
from sklearn import datasets
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
```

```python
dataset_renew = dataset.drop(columns=['RemovedTeeth', 'LastCheckupTime', 'ChestScan',
                       'HIVTesting', 'FluVaxLast12', 'PneumoVaxEver', 'TetanusLast10Tdap',
                       'HighRiskLastYear'])
dataset_renew.columns
```

```python
dataset_renew.describe(include='object')
```

```python
for col in dataset_renew.describe(include='object').columns:
    print('Column Name: ',col)
    print(dataset_renew[col].unique())
    print('-------------------------------------------------')
```

```python
dataset_renew['HadDiabetes'].replace({'No, pre-diabetes or borderline diabetes' : 'borderline',
                            'Yes, but only during pregnancy (female)' : 'during Pregnancy'} , inplace=True)

dataset_renew['HadDiabetes'].unique()
```

```python
dataset_renew['SmokerStatus'].replace({'Current smoker - now smokes some days' : 'current smoker(some days)',
                                    'Current smoker - now smokes every day' : 'current smoker(every day)'}, inplace=True)

dataset_renew['SmokerStatus'].unique()
```

```python
dataset_renew['ECigaretteUsage'].replace({'Not at all (right now)' : 'not at all',
                                        'Never used e-cigarettes in my entire life' : 'never',
                                        'Use them every day' : 'everyday',
                                        'Use them some days' : 'somedays'}, inplace=True)

dataset_renew['ECigaretteUsage'].unique()
```

```python
dataset_renew['RaceEthnicityCategory'].replace({'White only, Non-Hispanic' : 'white',
                                             'Black only, Non-Hispanic' : 'black',
                                             'Other race only, Non-Hispanic' : 'other Race',
                                             'Multiracial, Non-Hispanic' : 'multiracial'}, inplace=True)

dataset_renew['RaceEthnicityCategory'].unique()
```

```python
dataset_renew['AgeCategory'].replace({'Age 80 or older' : '80+'}, inplace=True)

for value in dataset_renew['AgeCategory'].unique()[1:]:
    value = str(value)
    if value != 'nan':
        cat_value = value.split(" ")
        dataset_renew['AgeCategory'].replace({value : cat_value[1]+"-"+cat_value[3]}, inplace=True)

dataset_renew['AgeCategory'].unique()
```

```python
dataset_renew['CovidPos'].replace({'Tested positive using home test without a health professional' : 'Yes'}, inplace=True)

dataset_renew['CovidPos'].unique()
```

```python
dataset_renew[dataset_renew.duplicated()]
```

```python
dataset_renew.info()
```

```python
categorical_columns = dataset_renew.columns
```

```python
for column in categorical_columns:
    dataset_renew[column] = pd.Categorical(dataset_renew[column])
    dataset_renew[column] = dataset_renew[column].cat.codes
dataset_renew.info()
```

```python
x = dataset_renew.drop(columns=['HadHeartAttack'])
y = dataset_renew['HadHeartAttack']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3,random_state=42)
```

```python
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
classifier = LogisticRegression(random_state = 0)
classifier.fit(x_train, y_train)

y_pred = classifier.predict(x_test)
print ("Accuracy : ", accuracy_score(y_test, y_pred))
```

```python
sv = SVC(kernel='linear')
sv.fit(x_train, y_train)
```

```python
train_score= sv.score(x_train, y_train)
print("Training Score:", train_score)
```

```python
test_score= sv.score(x_test, y_test)
print("Testing Score:", test_score)
```

```python
knn = KNeighborsClassifier(n_neighbors=8)
knn.fit(x_train, y_train)

y_pred = knn.predict(x_test)
print(accuracy_score(y_test, y_pred))
```
