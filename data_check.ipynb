```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.countplot(x = 'HadHeartAttack', data = data)
```

```python
data.loc[data['HadDiabetes'] == 'No, pre-diabetes or borderline diabetes', 'HadDiabetes'] = 'Pre'
data.loc[data['HadDiabetes'] == 'Yes, but only during pregnancy (female)', 'HadDiabetes'] = 'Pregnancy'
```

```python
plt.figure(figsize = (12, 12))

plt.subplot(331)
sns.countplot(data = data, x = 'HadAngina')

plt.subplot(332)
sns.countplot(data = data, x = 'HadStroke')

plt.subplot(333)
sns.countplot(data = data, x = 'HadAsthma')

plt.subplot(334)
sns.countplot(data = data, x = 'HadSkinCancer')

plt.subplot(335)
sns.countplot(data = data, x = 'HadCOPD')

plt.subplot(336)
sns.countplot(data = data, x = 'HadDepressiveDisorder')

plt.subplot(337)
sns.countplot(data = data, x = 'HadKidneyDisease')

plt.subplot(338)
sns.countplot(data = data, x = 'HadArthritis')

plt.subplot(339)
sns.countplot(data = data, x = 'HadDiabetes')

plt.tight_layout()
plt.show()
```
